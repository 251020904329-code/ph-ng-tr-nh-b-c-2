```python
n = int(input("Nhập số lượng n: "))

tong = 0

for i in range(n):
    x = float(input(f"Nhập số thứ {i+1}: "))
    tong += x

tbc = tong / n

print("Trung bình cộng là:", tbc)

```

```python
n = int(input("Nhập số lượng phần tử: "))

dem = 0
for i in range(n):
    x = int(input(f"Nhập phần tử thứ {i+1}: "))
    if x % 2 == 0:
        dem += 1

print("Số lượng số chẵn là:", dem)
```

```python
n = int(input("Nhập số lượng phần tử: "))

x = float(input("Nhập số thứ 1: "))
max = x
min = x

for i in range(2, n + 1):
    x = float(input(f"Nhập số thứ {i}: "))
    if x > max:
        max = x
    if x < min:
        min = x

print("Giá trị lớn nhất:", max)
print("Giá trị nhỏ nhất:", min)
```

```python
A = []

# Nhập 5 số
for i in range(5):
    x = float(input(f"Nhập số thứ {i+1}: "))
    A.append(x)

# Sắp xếp tăng dần
for i in range(4):
    for j in range(4 - i):
        if A[j] > A[j + 1]:
            A[j], A[j + 1] = A[j + 1], A[j]

print("Dãy số sau khi sắp xếp tăng dần:")
print(A)
```

```python
def giaithua(n):
    if n == 0 or n == 1:   # điều kiện dừng
        return 1
    else:
        return n * giaithua(n - 1)

n = int(input("Nhập n: "))
print(f"{n}! =", giaithua(n))
```
