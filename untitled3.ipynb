```python
import math

print("Giải phương trình bậc hai: a * x^2 + b * x + c = 0")

# 1. Nhập hệ số
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

# 2. Kiểm tra điều kiện của a
if a == 0:
    print("Vì a = 0, phương trình trở thành phương trình bậc nhất: b*x + c = 0")

    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm.")
        else:
            print("Phương trình vô nghiệm.")
    else:
        x = -c / b
        print("Phương trình có nghiệm duy nhất x =", x)

else:
    # 3. Tính delta
    delta = b**2 - 4*a*c
    print("Delta =", delta)

    # 4. Phân loại nghiệm
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có hai nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép:")
        print("x =", x)
    else:
        print("Phương trình vô nghiệm trong tập số thực (vì delta < 0).")
```

```python

n = int(input("Nhập một số nguyên dương n: "))

# Kiểm tra n hợp lệ
if n < 0:
    print("Vui lòng nhập một số nguyên dương!")
else:
    #  Tính tổng bằng vòng lặp for
    tong = 0
    for i in range(0, n+1):
        tong += i

    #  Hiển thị kết quả
    print("Tổng các số từ 0 đến", n, "là:", tong)
```

```python

n = int(input("Nhập một số nguyên dương n: "))

# Kiểm tra tính hợp lệ
if n <= 0:
    print("Vui lòng nhập một số nguyên dương!")
else:
    # 2. Tính tổng các số chẵn nhỏ hơn n
    tong = 0
    i = 0  # bắt đầu từ số chẵn đầu tiên

    while i < n:
        tong += i
        i += 2

    # 3. Hiển thị kết quả
    print("Tổng các số chẵn nhỏ hơn", n, "là:", tong)
```
