```python
import math

print("Giải phương trình bậc hai: a * x^2 + b * x + c = 0")

# 1. Nhập hệ số
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

# 2. Kiểm tra điều kiện của a
if a == 0:
    print("Vì a = 0, phương trình trở thành phương trình bậc nhất: b*x + c = 0")

    if b == 0:
        if c == 0:
            print("Phương trình có vô số nghiệm.")
        else:
            print("Phương trình vô nghiệm.")
    else:
        x = -c / b
        print("Phương trình có nghiệm duy nhất x =", x)

else:
    # 3. Tính delta
    delta = b**2 - 4*a*c
    print("Delta =", delta)

    # 4. Phân loại nghiệm
    if delta > 0:
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        print("Phương trình có hai nghiệm phân biệt:")
        print("x1 =", x1)
        print("x2 =", x2)
    elif delta == 0:
        x = -b / (2*a)
        print("Phương trình có nghiệm kép:")
        print("x =", x)
    else:
        print("Phương trình vô nghiệm trong tập số thực (vì delta < 0).")
```
