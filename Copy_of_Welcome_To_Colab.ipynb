```

n = int(input("Nhập số nguyên dương n: "))
tong = 0

for i in range(1, n + 1):
    if i % 2 == 0:
        tong += i

print("Tổng các số chẵn ≤ n là:", tong)
```

```
s = input("Nhập chuỗi: ")
words = s.split()

# Đếm số từ
so_tu = len(words)

# Tìm từ dài nhất
tu_dai_nhat = words[0]
for tu in words:
    if len(tu) > len(tu_dai_nhat):
        tu_dai_nhat = tu

print("So tu:", so_tu)
print("Tu dai nhat:", tu_dai_nhat)
```

```

nums = list(map(int, input("Nhập các số: ").split()))

# Loại bỏ phần tử trùng lặp
unique_nums = []
for x in nums:
    if x not in unique_nums:
        unique_nums.append(x)

# Sắp xếp tăng dần
unique_nums.sort()

# Tìm số lớn thứ hai
second_largest = unique_nums[-2]

print("Danh sách đã sắp sếp:", unique_nums)
print("Số lớn thứ 2 là:", second_largest)
```

```
# Nhập danh sách điểm
scores = list(map(int, input("Nhập các điểm: ").split()))

valid_scores = []
for s in scores:
    if 0 <= s <= 10:
        valid_scores.append(s)

# Tính điểm trung bình
average = sum(valid_scores) / len(valid_scores)

# Đếm số sinh viên đạt
passed = 0
for s in valid_scores:
    if s >= 5:
        passed += 1

print("Điểm trung bình:", average)
print("Số sinh viên đạt:", passed)
```
